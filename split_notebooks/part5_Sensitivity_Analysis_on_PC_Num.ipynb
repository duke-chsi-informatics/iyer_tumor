# Sankey diagram for clustering consistency

#### We want to use Sankey diagram to show that the clustering using 75 PC and 50 PC is not that different. If that is the case, each cluster of 50 PC clustering will be faithfully map to a cluster of 75 clustering.

```python
flows = {}
# can choose any source and target within the list umaps[]
source = umaps[13] 
target = umaps[14]
source_cluster_size = {}
# compute the flow from the i-th cluster of the source to the j-th cluster of the target
for i in set(source.obs['leiden']):
    source_cluster_size[i]=sum(source.obs['leiden']==i)
    for j in set(target.obs['leiden']):
        # compute the flow from i-th cluster of the source to the j-th cluster of the target
        flow = sum(target[source[source.obs['leiden']==i].obs.index,:].obs['leiden']==j)
        if flow>3 and flow>0.01*source_cluster_size[i]:
            flows[(i,j)]=flow
        if flow/source_cluster_size[i]>0.9:
            break
```

```python
source_cluster = []
target_cluster = []
flow_strength = []

# each arrow in the sankey diagram represents the flow with strength[i] from source_cluster_list[i] to target_cluster_list[i]
for flow in flows:
    if flows[flow]>4:
        source_cluster.append(int(flow[0]))
        target_cluster.append(int(flow[1]))
        flow_strength.append(flows[flow])
```

```python
labels = ['A'+x for x in map(str,range(len(set(source.obs['leiden']))))]+['B'+y for y in map(str,range(len(set(target.obs['leiden']))))]
# the source clusters are labeled 0-51 and the target cluster need to be labeled 52-103
adjusted_target_cluster = [x+len(set(source.obs['leiden'])) for x in target_cluster]
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = labels,
      color = "blue"
    ),
    link = dict(
      source = source_cluster, 
      target = adjusted_target_cluster,
      value = flow_strength
  ))])

fig.show()
```

# Similarity score for cluster assignments

#### We split the range into high number of PC used and low numer of PC used and see how similar results in these two ranges are. The higher the similarity the higher the reliability of the clustering.
#### To assess similarity between two clustering, we consider adjusted rand index, in general a adjusted rand index>0.8 means very high similarity
#### In each of the ranges, we compute all pairwise similarity. That is 
$$\mbox{adj rand index for low PC usage}=\{ARI(C_i,C_j)\}_{20\leq i<j\leq 40}$$
$$\mbox{adj rand index for high PC usage}=\{ARI(C_i,C_j)\}_{40< i<j\leq 70}$$
$$\mbox{adj rand index for high PC usage}=\{ARI(C_i,C_j)\}_{70< i<j\leq 100}$$
#### where $C_i$ denotes the result of cluster assignments for the cells when using $i$ PC's.
#### We then perform 1 tailed T-test for related samples and conclude that using large number of PC's will produce higher reliability of the clustering

```python
adj_rand_index_low_PC_usage = []
for i in range(0,31): 
    clustering_i = umaps[i].obs['leiden']
    for j in range(i+1,31):
        clustering_j = umaps[j].obs['leiden']
        adj_rand_index_low_PC_usage.append(adjusted_rand_score(clustering_i,clustering_j))
```

```python
adj_rand_index_medium_PC_usage = []
for i in range(30,61): 
    clustering_i = umaps[i].obs['leiden']
    for j in range(i+1,61):
        clustering_j = umaps[j].obs['leiden']
        adj_rand_index_medium_PC_usage.append(adjusted_rand_score(clustering_i,clustering_j))
```

```python
adj_rand_index_high_PC_usage = []
for i in range(60,91): 
    clustering_i = umaps[i].obs['leiden']
    for j in range(i+1,91):
        clustering_j = umaps[j].obs['leiden']
        adj_rand_index_high_PC_usage.append(adjusted_rand_score(clustering_i,clustering_j))
```

```python
adj_rand_index_low_PC_usage = pd.DataFrame(adj_rand_index_low_PC_usage)
mean = adj_rand_index_low_PC_usage.describe()[0]['mean']
std = adj_rand_index_low_PC_usage.describe()[0]['std']
print('For clusterings with low PC usage, the mean and standard deviation of adjusted rand indeces are: {}, {}'.format(round(mean,2),round(std,3)))
```

```python
adj_rand_index_medium_PC_usage = pd.DataFrame(adj_rand_index_medium_PC_usage)
mean = adj_rand_index_medium_PC_usage.describe()[0]['mean']
std = adj_rand_index_medium_PC_usage.describe()[0]['std']
print('For clusterings with medium PC usage, the mean and standard deviation of adjusted rand indeces are: {}, {}'.format(round(mean,2),round(std,3)))
```

```python
adj_rand_index_high_PC_usage = pd.DataFrame(adj_rand_index_high_PC_usage)
mean = adj_rand_index_high_PC_usage.describe()[0]['mean']
std = adj_rand_index_high_PC_usage.describe()[0]['std']
print('For clusterings with high PC usage, the mean and standard deviation of adjusted rand indeces are: {}, {}'.format(round(mean,2),round(std,3)))
```

```python
ttest_rel(adj_rand_index_high_PC_usage,adj_rand_index_medium_PC_usage)[1]
```

## Plotting rand indeces for clusterings with consecuitive numbers of PC 

#### We Compute and plot the rand indeces for clusterings with using $i$ and $i+1$ PC's ($10\leq i< 100$).

```python
adj_rand_index_neighboring_PC_usage = []
for i in range(0,90): 
    adj_rand_index_neighboring_PC_usage.append(adjusted_rand_score(umaps[i].obs['leiden'],umaps[i+1].obs['leiden']))
```

```python
plt.plot(range(10,100),adj_rand_index_neighboring_PC_usage)
plt.ylabel('rand index')
plt.xlabel('PC used')
plt.xlim(10,100)
plt.xticks(ticks=range(10,101,10))
plt.show()
```
