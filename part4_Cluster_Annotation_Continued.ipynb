```python
import scanpy as sc
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

```python
combined_tsne = sc.read('/scratch/combined_tsne.h5ad')
combined_umap = sc.read('/scratch/combined_umap.h5ad')
```

# TSNE plots with annotations

```python
tsne_cluster_names=['NKT/T','Microglial/Hematopoietic stem','Microglial/T','T/NK', 'B','undetermined','B ','T/NKT',
                    'Basal/Embryonic stem','NKT/T  ','Treg/T','NKT','Basal','NKT ','Mesenchymal stem/stem','Astrocyte/NKT',
                    'NKT/T ','Endothelial/Mesenchymal stem','Stem/Progenitor','B  ','NKT  ','Mast/Hematopoietic']
```

```python
print('Cluster Type')
for num,type in enumerate(tsne_cluster_names):
    print((num,type))
```

```python
combined_tsne.rename_categories('louvain', tsne_cluster_names)
sc.pl.tsne(combined_tsne, color='louvain', frameon=False)
sc.pl.tsne(combined_tsne, color='sample', frameon=False)
sc.pl.tsne(combined_tsne, color='group', frameon=False)
```

# UMAP plots with annotations

```python
umap_cluster_names=['NKT/T','Microglial/Hematopoietic stem','Microglial/T','T/NK', 'B','undetermined','B ','T/NKT',
                    'Basal/Embryonic stem','NKT/T  ','Treg/T','NKT','Basal','NKT ','Mesenchymal stem/stem','Astrocyte/NKT',
                    'NKT/T ','Endothelial/Mesenchymal stem','Stem/Progenitor','B  ','NKT  ','Mast/Hematopoietic']
```

```python
combined_umap.rename_categories('louvain', umap_cluster_names)
sc.pl.umap(combined_umap, color='louvain', title='UMAP', frameon=False)
sc.pl.umap(combined_umap, color='sample', frameon=False)
sc.pl.umap(combined_umap, color='group',frameon=False)
```
